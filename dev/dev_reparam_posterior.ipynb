# check piece by piece whether we compute what we want to

```python
from itertools import izip
# load trained network
net = helper.load(filename)[0]

n_dims = 1
n_outputs = 1
n_samples = 20

# generate data for querying correctness of computed quantities
ths = np.empty(n_samples)
xs = np.empty(n_samples)
dist = np.empty(n_samples)

for i in xrange(n_samples):

    ths[i] = prior['mu_a'] - 1.0
    while ths[i] < prior['mu_a'] or ths[i] > prior['mu_b']:
        ths[i] = prior['sim_prior']() 
    xs[i] = sim_likelihood(ths[i])
    dist[i] = calc_dist(xs[i], x_obs)
y = ths.copy().reshape(n_samples, n_dims)
    

# ask network for all quantities of interest
a, ms, Us, Ps, lprobs = net.eval_debug((xs.reshape(-1,1), ths.reshape(-1,1)))    

# compute target log probs as was done in original code (assuming P = U'*U)
diag_mask = np.eye(n_outputs)
ldetUs = [np.sum(np.sum(np.log(diag_mask * U), axis=2), axis=1) for U in Us]
lprobs_comps = [-0.5 * np.sum(np.sum((y-m).reshape([y.shape[0], 1, y.shape[1]]) * U, axis=2)**2, axis=1) + ldetU for m, U, ldetU in izip(ms, Us, ldetUs)]
lprobs_check = np.log(np.sum(np.exp(np.stack(lprobs_comps, axis=1) + np.log(a)), axis=1)) - (0.5 * n_outputs * np.log(2*np.pi))

print('comparing network-computed and ground-truth log-probs log(p(th | m(x), U(x))) (original-code style)')
print('[computed | ground-truth]')
print(np.hstack([lprobs.reshape(-1,1),lprobs_check.reshape(-1,1)]))


# compute target log probs under the reparametrisation (assuming P = U'*U + V'*V, with V=0 here)
ldetPs = [np.array([0.5 * np.log(np.linalg.det(U[j,:,:].T.dot(U[j,:,:]) + V.T.dot(V))) for j in range(n_samples)]) for U,V in izip(Us,net.Vs)]
lprobs_comps_U = [-0.5 * np.sum(np.sum((y-m).reshape([y.shape[0], 1, y.shape[1]]) * U, axis=2)**2, axis=1) for m, U in izip(ms, Us)]
lprobs_comps_V = [-0.5 * np.sum(np.sum((y-m).reshape([y.shape[0], 1, y.shape[1]]) * V.reshape([1, V.shape[0], V.shape[1]]), axis=2)**2, axis=1) for m, V in izip(ms, net.Vs)]
lprobs_comps = [lU + lV + ldetP for lU,lV,ldetP in izip(lprobs_comps_U,lprobs_comps_V,ldetPs)]
lprobs_check = np.log(np.sum(np.exp(np.stack(lprobs_comps, axis=1) + np.log(a)), axis=1)) - (0.5 * n_outputs * np.log(2*np.pi))

print('comparing network-computed and ground-truth log-probs log(p(th | m(x), U(x))) (reparametrized-style)')
print('[computed | ground-truth]')
print(np.hstack([lprobs.reshape(-1,1),lprobs_check.reshape(-1,1)]))

# also have a look at the logdets directly

print('comparing direct (tt.log.det(P)) and previous (sum(diag(U)) method to compute logdet(P)')
print('[computed | ground-truth]') 
print([np.hstack([ldetUs[i].reshape(-1,1),ldetPs[i].reshape(-1,1)]) for i in range(n_components)])

```

# test computation of logdet() using theano.scan() 

```python
import theano
import theano.tensor as tt
import numpy as np

n = 3
n_components = 2

Uarr = np.random.normal(size=(10,n,n))

def prec(U,V):
    return tt.dot(U.dimshuffle([1, 0]),U) + V.T.dot(V)
def ldetPrec(P):
    return .5 * tt.log(tt.nlinalg.Det()( P ))

U, V  = tt.tensor3('U'), np.zeros((n,n))

Ps = [theano.scan(fn=prec,
                  outputs_info  = None,
                  sequences = U,
                  non_sequences = V )[0] for i in xrange(n_components)]

ldetUVs = [theano.scan(fn=ldetPrec,
                       outputs_info  = None,
                       sequences = Ps[i])[0] for i in xrange(n_components)]
```
